```python
import pandas as pd
import numpy as np
```

```python
df2=pd.read_csv("C:/Users/Thrilokboss/Documents/IVBCADA/PRODIGY/data.csv")
df2
```

```python
data=pd.DataFrame(df2)
```

```python
Remove_Unwanted=data.loc[:,'Indicator Name':'Indicator Code']
```

```python
data.drop(columns=Remove_Unwanted,inplace=True)
```

```python
data
```

```python
print(data.isnull().sum())
```

```python
#Replacing the missing values to the mean of the particular column
numeric_columns=data.select_dtypes(include=["number"]).columns
print(numeric_columns)
data[numeric_columns]=data[numeric_columns].fillna(data[numeric_columns].mean())
```

```python
print(data.isnull().sum())
```

```python
data['Avg_Population'] = data.iloc[:,2:-1].mean(axis=1)
data.head(10)
```

```python
sort=data.sort_values(by="Avg_Population",ascending=False)
Top10_Populated_Countries=sort.head(10)
Top10_Populated_Countries
```

```python
df1=pd.read_csv("C:/Users/Thrilokboss/Documents/IVBCADA/PRODIGY/dataa.csv")
df1.drop("Code",axis=1,inplace=True)
df1.columns
df1.rename(columns={'Name':'Country Name'},inplace=True)
```

```python
countries = pd.merge(data, df1, on='Country Name', how='inner')
sort=countries.sort_values(by="Avg_Population",ascending=False)
Top_10_Populated_Country=sort
Top_10_Populated_Country
```

```python
Top_10_Populated_Countries = pd.DataFrame(Top_10_Populated_Country.loc[:,"1960":'Avg_Population'].astype(int))
Top_10_Populated_Countries
```

```python
final_data = pd.concat([countries['Country Name'],Top_10_Populated_Countries],axis=1 )
Ready_data=final_data.sort_values(by="Avg_Population",ascending=False)
Populated_Countries = Ready_data
Populated_Countries
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a horizontal bar plot
plt.figure(figsize=(15, 5))
sns.barplot(data=Populated_Countries[0:10], y='Country Name', x='Avg_Population', orient='h')

# Set the title and labels
plt.title('Populated Countries')
plt.xlabel('Average Population')
plt.ylabel('Country Name')

# Display the plot
plt.show()
```

```python
# Extract the data for the five countries
china_data = Populated_Countries[Populated_Countries['Country Name'] == 'China'][["2018", "2019", "2020", "2021", "2022"]]
india_data = Populated_Countries[Populated_Countries['Country Name'] == 'India'][["2018", "2019", "2020", "2021", "2022"]]
us_data = Populated_Countries[Populated_Countries['Country Name'] == 'United States'][["2018", "2019", "2020", "2021", "2022"]]
indo_data = Populated_Countries[Populated_Countries['Country Name'] == 'Indonesia'][["2018", "2019", "2020", "2021", "2022"]]
braz_data = Populated_Countries[Populated_Countries['Country Name'] == 'Brazil'][["2018", "2019", "2020", "2021", "2022"]]


# Create a line chart
plt.figure(figsize=(15, 5))
plt.plot(china_data.columns, china_data.values[0], label='China')
plt.plot(india_data.columns, india_data.values[0], label='India')
plt.plot(us_data.columns, us_data.values[0], label='United States')
plt.plot(indo_data.columns, indo_data.values[0], label='Indonesia')
plt.plot(braz_data.columns, braz_data.values[0], label='Brazil')

# Set the title and labels
plt.title('Population of Top_5_Populated Countries from 2018 to 2022')
plt.xlabel('Year')
plt.ylabel('Population')

# Display the legend
plt.legend()

# Display the plot
plt.show()
```
